```python
import json
import keras
import keras.backend as kb
import pandas as pd
import requests
import tensorflow as tf

from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('data/dow_jones_index.data')
```

```python
dataset
```

```python
dataset['mid_range'] = dataset.apply(lambda x: (float(x.high.replace('$','')) + float(x.low.replace('$','')))/2, axis=1)
```

```python
date_encoder = LabelEncoder()
```

```python
date_encoder = date_encoder.fit(dataset.date.values)
```

```python
train_dataset = pd.DataFrame(columns=['X_train', 'Y_train'])
train_dataset.X_train = date_encoder.transform(dataset.date.values)
train_dataset.Y_train = dataset['mid_range'].values
```

```python
#Creating Feature Columns with Tensorflow Objects
def create_feature_columns():

    categorical_columns_with_vocabulary_list = tf.feature_column.indicator_column(tf.feature_column.categorical_column_with_vocabulary_list(key='date', vocabulary_list=date_encoder.classes_))

    feature_columns = {}
    feature_columns.update({'date': categorical_columns_with_vocabulary_list})


    return feature_columns


# Creating the input function for the training of the model with Tensorflow
def build_input_fn_train():
    
    return tf.estimator.inputs.pandas_input_fn(
        x= pd.DataFrame(train_dataset.X_train.values, columns=['date']).astype(str),
        y= train_dataset.Y_train.astype(float),
        batch_size=16,
        num_epochs=500,
        shuffle=True,
        num_threads=8) 


# Creating the input function for the predictions of the model with Tensorflow
def build_input_fn_test():

    return tf.estimator.inputs.pandas_input_fn(
        x= pd.DataFrame(train_dataset.X_train.values, columns=['date']).astype(str),
        shuffle=False)

#Creating input receiver function for model export
def create_input_receiver_serving():
    def input_receiver():

        inputs = {}

        inputs['date'] = tf.placeholder(shape=[None], dtype=tf.string)

        features_dict = {key: tf.expand_dims(tensor, -1)for key, tensor in inputs.items()}

        from pprint import pprint
        pprint("****************************************")
        pprint("serving_feature_columns")
        pprint(features_dict)
        pprint("****************************************")

        return tf.estimator.export.ServingInputReceiver(
            features=features_dict,
            receiver_tensors= inputs
        )
        
    return input_receiver
```

```python
# Creating Optimizer
def create_optimizer():
    return lambda: tf.train.AdamOptimizer(tf.train.exponential_decay(learning_rate=0.0001,global_step=tf.train.get_global_step(),decay_steps=8888,decay_rate=0.98))
```

```python
def create_regressor():

    estimator = tf.estimator.DNNLinearCombinedRegressor(
        model_dir='/saved_modelsss/',
        dnn_feature_columns=create_feature_columns().values(),
        dnn_hidden_units=[32,32],
        dnn_optimizer=create_optimizer(),
        dnn_activation_fn=tf.nn.relu,
    )
    
    return estimator
```

```python
estimator = create_regressor()
```

```python
estimator.train(input_fn=build_input_fn_train())
```

```python
predictions = estimator.predict(input_fn=build_input_fn_test())
```

```python
train_dataset['results'] = [w['predictions'][0] for w in predictions]
```

```python
estimator.export_savedmodel('saved_model', serving_input_receiver_fn=create_input_receiver_serving())
```
